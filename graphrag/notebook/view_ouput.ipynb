## Community Report

```python
import json
from textwrap import fill
```

```python
summary = df_report[df_report['human_readable_id'] == 12]['full_content'].values[0]
print(summary)
```

```python
def print_report(id=183):
    summary = df_report[df_report['human_readable_id'] == id]['full_content_json']
    text = summary.values[0]
    content = json.loads(text)

    print(content["title"])
    print()
    print(fill(content['summary'], width=100))
    print()

    print("## Findings:")
    for finding in content['findings']:
        print(f"- {finding['summary']}")
        print(fill(finding['explanation'], width=100))
        print()

def list_reports():
    for idx, row in df_report.iterrows():
        print(f"{row['human_readable_id']}: {row['title']}")
```

```python
list_reports()
```

```python
print_report(4)
```

## Entity & Relationships

```python
df_entity = pd.read_parquet("../output/entities.parquet")
df_entity.head()
```

```python
len(df_entity)
```

```python
df_entity["type"].unique()
```

```python
def print_entity(id):
    entity = df_entity[df_entity['human_readable_id'] == id]
    print(f"{entity['title'].values[0]}, ({entity['type'].values[0]})")
    print()

    print(fill(entity['description'].values[0], width=100))
    print()
    print(f"Source: {entity['text_unit_ids'].values[0]}")
```

```python
print_entity(4)
```

### Relationship

```python
df_relationship = pd.read_parquet("../output/relationships.parquet")
df_relationship.head()
```

```python
def print_relationship(id=4):
    rel = df_relationship[df_relationship['human_readable_id'] == id]
    print(f"{rel['source'].values[0]}->{rel['target'].values[0]}")
    print(rel['description'].values[0])
    print()
    print(f"Source: {rel['text_unit_ids'].values[0]}")
```

## Text Unit

```python
df_text = pd.read_parquet("../output/text_units.parquet")
df_text.head()
```

```python
text_unit_id = '243941f52781cd785f52d6af493c2efe8dd2948ab29c8a6a750beb0e22b783bee0a7df518c91f26de4a408c62993ac612ed3f18fc9d0fd7ed9699ce315e67dbb'
df_text[df_text['id'] == text_unit_id]
```

```python
def print_text_unit(text_unit_id):
    text_unit = df_text[df_text['id'] == text_unit_id]
    print(text_unit['document_ids'].values[0])
    doc_ids = text_unit['document_ids'].values[0]
    for id in doc_ids:
        doc = df_doc[df_doc['id'] == id]
        print(f"- {doc['title'].values[0]}")
    print()
    print(fill(text_unit['text'].values[0], width=100))
```

```python
print_text_unit(text_unit_id)
```
