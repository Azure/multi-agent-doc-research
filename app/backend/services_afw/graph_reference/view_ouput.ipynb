## Community Report

```
2. 정치적 및 지정학적 리스크 요약 및 은행 산업에 대한 영향
프랑스는 최근 몇 년간 정치적 불안정과 사회적 반발을 경험하고 있습니다. 2023년과 2024년 연금 및 예산 개혁이 의회 승인 없이 헌법 권한으로 추진되면서 노란 조끼 시위 등 사회적 긴장이 고조되었고, 2022년 총선에서 마크롱 대통령의 연립정부가 의회 과반을 상실하는 등 입법 추진에 어려움을 겪고 있습니다. 이러한 정치적 불안정은 금융시장에 불확실성을 가중시키며, 은행 산업의 규제 환경과 신용 리스크에 영향을 미칠 가능성이 있습니다 [Data: Reports (183)].
```

````python
import json
from textwrap import fill
````

````python
summary = df_report[df_report['human_readable_id'] == 183]['full_content'].values[0]
print(summary)
````

````python
def print_report(id=183):
    summary = df_report[df_report['human_readable_id'] == id]['full_content_json']
    text = summary.values[0]
    content = json.loads(text)

    print(content["title"])
    print()
    print(fill(content['summary'], width=100))
    print()

    print("## Findings:")
    for finding in content['findings']:
        print(f"- {finding['summary']}")
        print(fill(finding['explanation'], width=100))
        print()

def list_reports():
    for idx, row in df_report.iterrows():
        print(f"{row['human_readable_id']}: {row['title']}")
````

````python
list_reports()
````

````python
print_report(183)
````
